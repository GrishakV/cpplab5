Построить график изменения ускорения свободного падения от высоты Н над поверхностью Земли.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

x = [0, 10, 100, 1000, 10000, 100000, 1000000]

m1 = 12
m2 = 5
fi = [45, 2, 27.85]
fi = fi[0] + (fi[1] / 60) + (fi[2] / 3600)
fi = math.radians(fi)
y = np.array([(9.780318 * (1 + 0.005302 * (math.sin(fi))**2 - 0.000006 * (math.sin(fi))**2) - 0.000003086 * i) for i in x])
print(y)
plt.title('Зависимость ускорения свободного падения от высоты над поверхностью Земли')
plt.xlabel("Высота")
plt.ylabel("Ускорение свободного падения")
plt.grid(True)

plt.plot(x, y, 'o-g', alpha=0.7, label="first", lw=5, mec='r', mew=2, ms=10)
```
