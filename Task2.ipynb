**Создайте столбчатые диаграммы для отображения некоторых сравнительных показателей экономического развития России. Уровень экономического развития определяют:**

* **уровень ВВП;**
* **ВНП на душу населения**

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
AutoMinorLocator)
import matplotlib.colors as mcolors
```

```python
USA = [52615, 54360, 55805]
JAP = [36787, 37442, 38054]
UK = [38724, 40163, 41159]
KAN = [43680, 44990, 45553]
RUS = [25987, 26138, 25411]

width = 0.1

x = np.arange(3)

fig, ax = plt.subplots()
rects1 = ax.bar(x - width*2, USA, width, label='USA')
rects2 = ax.bar(x - width, JAP, width, label='JAP')
rects3 = ax.bar(x , UK, width, label='UK')
rects4 = ax.bar(x + width, KAN, width, label='KAN')
rects5 = ax.bar(x + width*2, RUS, width, label='RUS')

ax.set_title('Уровень ВВП')
ax.set_xticks(x)
ax.set_xticklabels(('2013', '2014', '2015'))

ax.legend()
```

```python
groups = ['USA', 'JAP', 'UK', 'KAN', 'RUS']
counts = [60200, 44850, 42560, 46070, 24890]

plt.title('ВНП стран мира на душу населения')
plt.barh(groups, counts)
```
