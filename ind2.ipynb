```python
groups = ['Java', 'Python', 'C#', 'JavaScript', 'TypeScript', 'Groovy', 'C',
     'C++', 'PHP', 'Ruby']

counts = [39.3, 21.3, 18.9, 12.3, 3.3, 1.6, 0.8, 0.8, 0.8, 0.8]

plt.title('Популярность языков программировани')
plt.xlabel("Популярность в %")
plt.ylabel("Языки")
plt.barh(groups, counts)
```
